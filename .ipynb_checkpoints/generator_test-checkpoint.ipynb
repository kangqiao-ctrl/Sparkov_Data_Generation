```python
cate = ['ssn','cc_num','first','last','street','zip','state','dob','acct_num','profile',
                            'trans_num','unix_time']
new_col = [x for x in df.columns.to_list() if x not in cate ]
df_clean = df[new_col]
```

```python
df_clean['gender'] = df_clean['gender'].astype('category')
df_clean['city'] = df_clean['city'].astype('category')
df_clean['job'] = df_clean['job'].astype('category')
df_clean['city'] = df_clean['city'].astype('category')
df_clean['category'] = df_clean['category'].astype('category')
df_clean['is_fraud'] = df_clean['is_fraud'].astype('category')
df_clean['merchant'] = df_clean['merchant'].astype('category')
```

```python
df_clean['city_pop'] = df_clean['city_pop'].astype('int16')
```

```python
cat_columns = df_clean.select_dtypes(['category']).columns
df_clean[cat_columns] = df_clean[cat_columns].apply(lambda x: x.cat.codes)
```

```python
df_clean.info()
```

```python
df_clean['trans_date'] = pd.to_datetime(df_clean['trans_date']).dt.day
df_clean['trans_time'] = pd.to_datetime(df_clean['trans_time']).dt.hour
```

```python
df_clean['trans_date'].nunique()
```

```python
df_clean['trans_time'].nunique()
```

```python
df_min = df_clean.sample(n = 50000)
```

```python
torch.cuda.get_device_name(0)
```

```python
torch.cuda.init()
torch.cuda.is_initialized()
```

```python
torch.multiprocessing.set_start_method('spawn', force=True)
```

```python
model = SAM(njobs=2, gpus=1,nruns=4,mixed_data = True,verbose = True)
output = model.predict(df_min)
```

```python
a = nx.draw_networkx(output, font_size=12)
plt.show()
am = nx.adjacency_matrix(output).todense()
am
```

```python
df_min.to_csv("df_min.csv",index=None)
```

```python
amd = pd.DataFrame(am)
amd.columns = df_min.columns
amd.index =df_min.columns
amd.to_csv("amd.csv")
```
